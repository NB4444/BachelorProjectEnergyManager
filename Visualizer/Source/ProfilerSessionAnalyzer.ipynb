## Profiler Sessions Overview

Below all profiler sessions are shown.

```python
# Load all profiler sessions
profiler_sessions = ProfilerSession.load_all(database)

# Visualize the profiler session data
ProfilerSession.table(profiler_sessions).plot()
```

### Visualize a Profiler Session Analysis

Define the selection criteria for the profiler sessions you wish to analyze below.

```python
selected_profiler_sessions = [profiler_session for profiler_session in profiler_sessions if profiler_session.label == "Matrix Multiply" and profiler_session.profile["matrixAWidth"] == str(32 * 30)][:100]
# selected_profiler_sessions = [profiler_session for profiler_session in profiler_sessions if profiler_session.label == "KMeans" and profiler_session.profile["file"].endswith("100")][:100]
# selected_profiler_sessions = profiler_sessions

# Visualize the profiler sessions data
ProfilerSession.table(selected_profiler_sessions).plot()
```

#### Summary

These are the specifications of the systems with which the profiler sessions were performed.

```python
# Generate summary
pprint.pprint(Containers.unique_dict_of_lists(Containers.merge_dictionaries([profiler_session.summary for profiler_session in selected_profiler_sessions])), compact=True)
```

#### Monitor Data

These are the results returned by the monitors that ran during the profile executions.

```python
display.display(Containers.merge_tables([monitor_datum.table().pandas_table for profiler_session in selected_profiler_sessions for monitor_datum in profiler_session.monitor_data]))

# Show the results dictionary
global_monitor_results = Containers.unique_dict_of_lists(Containers.merge_dictionaries([variables for profiler_session in selected_profiler_sessions for monitor_datum in profiler_session.monitor_data for _, variables in monitor_datum.namespaced_monitor_data.items()]))
# pprint.pprint(global_monitor_results, compact=True)

for title, values in global_monitor_results.items():
    if title.endswith("coreClockRate") or title.endswith("coreUtilizationRate"):
        HistogramPlot(title=title, values=values, bins=30).plot()
```
